```python
unique_script = df['script'].unique()

# Define a color palette, ensuring there's a color for each unique genus
palette = sns.color_palette("husl", len(unique_script))
script_color_map = dict(zip(unique_script, palette))

# Set up the figure
plt.figure(figsize=(16, 16))

# Plot a bar plot for each genus group
for i, script in enumerate(unique_script, start=1):
    # Create a subplot for each genus
    plt.subplot(5, 4, i)
    
    # Filter the dataframe for the current genus
    subset = df[df['script'] == script].sort_values('mgsm_cot_llama_eng_instr_7b', ascending=False)
    
    # Create the bar plot with a specific color for the genus
    sns.barplot(data=subset, x='Language', y='mgsm_cot_llama_eng_instr_7b', color=script_color_map[script])
    
    
    # Set title and adjust x-tick labels for readability
    plt.title(script)
    plt.ylim(0,25)
    plt.xticks(rotation=45)
    plt.ylabel('Accuracy (%)')
    plt.xlabel('Language')

plt.suptitle('Accuracy on MGSM with Llama-2-7b-chat per script')
plt.tight_layout()
plt.show()
```
