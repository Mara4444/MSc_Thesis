```python
df1 = pd.DataFrame({'Language' : xcopa_langs + mt_langs_xcopa,
                    'Native' : xcopa_basic_llama_native_instr + xcopa_mt_basic_llama_native_instr,
                    'English' : xcopa_basic_llama_eng_instr + xcopa_mt_basic_llama_eng_instr
                    })

corpus = pd.read_csv('../EDA/training_corpus_llm.csv',sep=';')
corpus = corpus[['Language','% in Llama-2-7b']]

target_langs = xcopa_langs + mt_langs_xcopa
   
corpus = corpus[corpus['Language'].isin(target_langs)]

missing_langs = [lang for lang in target_langs if lang not in corpus['Language'].unique()]

missing_langs_df = pd.DataFrame(missing_langs, columns=['Language'])

for col in corpus.columns[1:]:  
    missing_langs_df[col] = np.nan

corpus = pd.concat([corpus, missing_langs_df], ignore_index=True)     

df = pd.merge(corpus, df1, on='Language')
df['Seen'] = np.where(df['% in Llama-2-7b'] > 0, 1, 0)

high_resource = ['English'] 
medium_resource = ["Chinese",'Dutch', 'German', 'French', 'Italian', 'Russian', 'Spanish', 'Swedish','Japanese', ]   
low_resource = ["Bulgarian", "Catalan", "Czech", "Danish", "Finnish", "Hungarian", "Indonesian", "Norwegian", "Romanian", "Serbian", "Slovenian", "Croatian",'Korean', 'Polish', 'Portuguese', 'Vietnamese', 'Ukrainian']
unseen = ["Afrikaans", "Balinese", "Bosnian", "Cantonese", "Estonian", "Haitian","Javanese", "Malay", "Quechuan", "Slovak", "Swahili", "Turkish", "Zulu",'Basque']
unseen_script = ["Arabic", "Armenian", "Belarusian", "Bengali", "Burmese", "Greek", "Hebrew", "Hindi", "Khmer", "Lao", "Maithili", "Malayalam", "Marathi", "Nepali", "Tagalog", "Tamil", "Telugu", "Thai", "Tibetan", "Urdu"]

def get_coverage(language):
    if language in high_resource:
        return "High"
    elif language in medium_resource:
        return "Medium"
    elif language in low_resource:
        return "Low"
    elif language in unseen:
        return "Unseen"
    elif language in unseen_script:
        return "Unseen script"
    else:
        return None

# Apply function to the Language column
df['Coverage'] = df['Language'].apply(get_coverage)

df['Delta'] = df['English'] - df['Native']

colors = ["#66c2a5", "#e78ac3", "#8da0cb", "#fc8d62", "#a6d854"] 
category_order = ['High', 'Medium', 'Low', 'Unseen', 'Unseen script']

# Map the colors to the categories
color_palette = dict(zip(category_order, colors))

plt.figure(figsize=(10, 5)) 

# First subplot for XCOPA
plt.subplot(1, 2, 1)  
sns.barplot(x='Coverage', y='Native', hue='Coverage', data=df, order=category_order, capsize=.1, palette=color_palette, errorbar='sd', legend=False)
plt.title('XCOPA')
plt.xlabel('')
plt.ylabel('Avg. performance (%)')
plt.xticks(rotation=45)
plt.yticks(np.arange(0, 120, 20))
plt.grid(False)

plt.figure(figsize=(10, 5)) 

# First subplot for XCOPA
plt.subplot(1, 2, 1)  
sns.barplot(x='Coverage', y='English', hue='Coverage', data=df, order=category_order, capsize=.1, palette=color_palette, errorbar='sd', legend=False)
plt.title('XCOPA')
plt.xlabel('')
plt.ylabel('Avg. performance (%)')
plt.xticks(rotation=45)
plt.yticks(np.arange(0, 120, 20))
plt.grid(False)


```
