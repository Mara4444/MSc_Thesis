```python
df = pd.DataFrame({'Language' : xcopa_langs_abv + mt_langs_abv,
                    'Native' : xcopa_basic_bloomz_native_instr + xcopa_mt_basic_bloomz_native_instr,
                    'English' : xcopa_basic_bloomz_eng_instr + xcopa_mt_basic_bloomz_eng_instr
                    })

high_resource = ['ar', 'ca', 'zh', 'en', 'fr', 'id', 'pt', 'es', 'vi']
medium_resource = ['bn', 'hi', 'ml', 'ta', 'ur','eu'] 
low_resource = ['mr', 'ne', 'sw', 'te', 'zu']
unseen = [
    "af", "ban", "bs", "bg", "be","my", 
    "yue", "cs", "da", "de", "et", 
    "fi", "ht", "hr", "el", "hu", 
    "it", "ja", "jv", "mai", "ms", 
    "pl", "nl", "nn", "qu", "ro", 
    "ru", "sk", "sl", "sr","sv", "tl", 
    "th", "tr", "uk"
]

def get_coverage(language):
    if language in high_resource:
        return "High"
    elif language in medium_resource:
        return "Medium"
    elif language in low_resource:
        return "Low"
    elif language in unseen:
        return "Unseen"
    else:
        return None

# Apply function to the Language column
df['Coverage'] = df['Language'].apply(get_coverage)

df['Delta'] = df['English'] - df['Native']

category_order = ['High', 'Medium', 'Low', 'Unseen']

fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)



# Plot Native performance
sns.barplot(ax=axes[0], x='Coverage', y='Native', hue='Coverage', data=df, order=category_order, capsize=.1, palette="Set2", errorbar='sd', dodge=False)
axes[0].set_title('Native instructions',fontsize=14)
axes[0].set_xlabel('')
axes[0].set_ylabel('Avg. performance (%)',fontsize=14)
axes[0].tick_params(axis='x', rotation=0)
axes[0].set_ylim(0, 100)
axes[0].grid(False)

# Plot English performance
sns.barplot(ax=axes[1], x='Coverage', y='English', hue='Coverage', data=df, order=category_order, capsize=.1, palette="Set2", errorbar='sd', dodge=False)
axes[1].set_title('English instructions',fontsize=14)
axes[1].set_xlabel('')
axes[1].set_ylabel('')
axes[1].tick_params(axis='x', rotation=0)
axes[1].set_ylim(0, 100)
axes[1].grid(False)

# Plot Delta performance
sns.barplot(ax=axes[2], x='Coverage', y='Delta', hue='Coverage', data=df, order=category_order, capsize=.1, palette="Set2", errorbar=None, dodge=False)
axes[2].set_title(r'Performance $\Delta$',fontsize=14)
axes[2].set_xlabel('')
axes[2].set_ylabel('')
axes[2].tick_params(axis='x', rotation=0)
axes[2].set_ylim(0, 100)
axes[2].grid(False)

plt.rc('xtick', labelsize=15)

plt.tight_layout()
plt.show()

print(df.groupby('Coverage')['Delta'].mean())

```
